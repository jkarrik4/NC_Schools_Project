```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.wearecms.com/apps/pages/allschools'
browser.visit(url)
```

```python
html = browser.html
page_content = BeautifulSoup(html, 'html.parser')
```

```python
schoolData = page_content.findAll('div', class_='page-block page-block-text')
schoolData
```

```python
schoolNames = page_content.findAll("div", class_= "collapsible-content")
elementarySchool = schoolNames[1]
elementary= elementarySchool.findAll("a")
elementary[1].text
```

```python
schoolNames = page_content.findAll("div", class_= "collapsible-content")
middleSchool = schoolNames[2]
middle= middleSchool.findAll("a")
middle[2].text
```

```python
schoolNames = page_content.findAll("div", class_= "collapsible-content")
highSchool = schoolNames[3]
high= highSchool.findAll("a")
high[3].text
```

```python
elementaryList = []
for i in range(len(elementary)):
    text= elementary[i].text
    elementaryList.append(text)

elementaryList
```

```python
middleList = []
for i in range(len(middle)):
    text= middle[i].text
    middleList.append(text)

middleList
```

```python
highList = []
for i in range(len(high)):
    text= high[i].text
    highList.append(text)

highList
```

```python
schooldf = pd.DataFrame({"Elementary": pd.Series(elementaryList),
             "Middle" : pd.Series(middleList),
             "High": pd.Series(highList)})
schooldf
```
